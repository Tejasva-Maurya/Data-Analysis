# **Mall Customer Segmentation by Cluster Analysis Technique of Machine learning**

**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

import warnings 
warnings.filterwarnings("ignore")
```

**Reading Dataset**

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df.columns
```

```python
df.shape
```

**Variable Identification**

```python
df.info()
```

```python
df.describe()
```

```python
# checking null values
print(df.isnull().sum())
```

```python
# selecting the column for clusterning
features = df[["Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
# number of clusters
num_clusters = 5
```

```python
# Creating an instance of K-means model
kmeans = KMeans(n_clusters= num_clusters , random_state = 42)
```

```python
kmeans.fit(features)
```

```python
# Getting labels of cluster
cluster_labels = kmeans.labels_
cluster_labels
```

```python
df["Cluster"] = cluster_labels
```

```python
# Visualizing the clusters

plt.figure(figsize=(12,10))
sns.scatterplot(x= "Annual Income (k$)" , y = "Spending Score (1-100)", hue = "Cluster" , data = df)
plt.title("Customer Segmentation by K-means Clustering")
plt.show()
```

**Cluster Centroids**

```python
centroids = kmeans.cluster_centers_

plt.figure(figsize=(12,10))
sns.scatterplot(x= "Annual Income (k$)" , y = "Spending Score (1-100)", hue = "Cluster" , data = df)
plt.scatter(centroids[:,0], centroids[:,1], marker= "X" , s=100, color = "Green", label = "Centroids")
plt.title("Customer Segmentation by K-means Clustering with Centroids")
plt.legend()
plt.show()
```

**Cluster Silhouette Analysis:**

```python
from sklearn.metrics import silhouette_samples, silhouette_score

silhouette_vals = silhouette_samples(features, cluster_labels)
silhouette_avg = silhouette_score(features, cluster_labels)

plt.figure(figsize=(8, 6))
sns.histplot(silhouette_vals, bins=30)
plt.axvline(x=silhouette_avg, color='red', linestyle='-', label='Average Silhouette Score')
plt.title('Silhouette Analysis')
plt.xlabel('Silhouette Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Feature Distribution:**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Annual Income (k$)', data=df)
plt.title('Distribution of Annual Income by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Annual Income (k$)')
plt.show()
```

**Dimensionality Reduction Visualization :**

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
tsne_data = tsne.fit_transform(features)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=tsne_data[:, 0], y=tsne_data[:, 1], hue=cluster_labels, palette='tab10')
plt.title('t-SNE Visualization of Clusters')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend()
plt.show()
```
